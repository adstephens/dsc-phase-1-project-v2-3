### Data Analysis


To see the size before I begin to merge data to make sure data isn't lost `df.shape` is used. To see what columns exist and null data within `df.isna().sum()` is utilized. By dividing the missing data sum by size will give you the percentage of missing data; significant missing data will be replaced or culled on a case by case bases. 

Now to look at the columns present in each table in anticipation of merges; replace or delete null values and delete unnecessary columns.

```python
# Formula to shorten time and show
# the 'Shape', 'Null Sum' and 'Null Percentage' 
# per column within a dataframe.
def spread(df):
    PercentNull = (df.apply(lambda x: x.isna().sum()
                   / len(df))) * 100 
    return print(f'***Shape:***', df.shape,
                 f'***Null Sum Per Column:***', df.isna().sum(),
                 f'***Null Percentage within Columns:***',
                 (round(PercentNull, 2)))
```

By using the function 'spread', we can see that this data frame does not contain any missing data as shown above. Now, lets look to see if there are duplicates.

```python
# To check and see the amount of 
# data duplicated wtihin column
def dups(column,df):
    dup = df[df.duplicated(keep=False)]
    return dup.sort_values([column])
```

```python
# To see duplicates of specific dataframe columns
def CDup(column,df):
    dupOfColumn = df.duplicated(subset = [column]).sum()
    return dupOfColumn, dupOfColumn.sort_values([column])
```

```python
 # To check and see the amount of 
# data duplicated 
def dupS(df):
        dup = df.duplicated(keep=False)
        sumOfDups = dup.values.sum()
        boolenOfDups = dup.value_counts()
        return sumOfDups, boolenOfDups 
```

Identifying Return on Investment (ROI)

Return on investment specifically looks at the dollar amount earned for the amount invested in a project. Instead of looking at overall profit, it looks at the specific benefit from the project divided by the costs. Here we will use `production_budget`, `domestic_gross`, `worldwide_gross`. 


```python
MMSEIdf['ROI'] = (round(
    (MMSEIdf['worldwide_gross'])
    / MMSEIdf['production_budget']
    *100, 1)
)
MMSEIdf.head()
```

```python
# Gross Profit Margin = (Total Profit-Total Costs)/100
MMSEIdf['GPM'] = (round(
    (MMSEIdf['worldwide_gross'] - MMSEIdf['domestic_gross'])
    - MMSEIdf['production_budget']
    /100, 1)
)
MMSEIdf.tail()
```
